```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x="Date",
    y="Estimated Unemployment Rate (%)",
    data=df
)

plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x="Region",
    y="Estimated Unemployment Rate (%)",
    data=df
)

plt.xticks(rotation=90)

plt.title("Unemployment Rate by Region")

plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)

plt.title("Correlation Matrix")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Matrix")

plt.show()
```

```python
sns.lineplot(
    x="Date",
    y="Estimated Unemployment Rate (%)",
    hue="Region",
    data=df
)

plt.show()
```

Insights
	1.	The analysis shows that unemployment rates vary significantly across different regions. Some regions experience higher unemployment compared to others, which may be due to differences in economic development and employment opportunities.
	2.	The visualization of unemployment rates over time indicates noticeable fluctuations in employment levels. These fluctuations may be influenced by economic conditions and seasonal employment patterns.
	3.	During the COVID-19 pandemic period, unemployment rates increased in many regions. This suggests that lockdowns and economic disruptions had a major impact on employment.
	4.	The correlation analysis shows relationships between unemployment rate, labour participation rate, and employment levels. These relationships help understand how workforce participation affects unemployment trends.
